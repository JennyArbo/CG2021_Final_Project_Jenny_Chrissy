# Creating our database

```python
#Importing the module and creating a new connection
import psycopg2
con = psycopg2.connect("user=postgres password='PASSWORD'");

from psycopg2.extensions import ISOLATION_LEVEL_AUTOCOMMIT

 

#Connecting to Postgresql

con.set_isolation_level(ISOLATION_LEVEL_AUTOCOMMIT);

#Initializing a cursor and creating a new database for this project

cursor          = con.cursor();

name_Database   = "national_parks_biodiversity";


sqlCreateDatabase = "create database "+name_Database+";"

 

cursor.execute(sqlCreateDatabase);

con.close()
```

# Creating our plant tables

## Plant detail table

```python
#initiate the connection & set the cursor
con = psycopg2.connect("dbname=national_parks_biodiversity user=postgres")
cur = con.cursor()

#Create table and identify column headers, setting primary key to ID for all tables
cur.execute(
    '''
    CREATE TABLE plant_species_details8 (
        id int,
        Park_name text,
        Scientific_name text,
        Category text,
        Plant_Order text,
        Family text,
        Record_status text,
        plant_animal_boolean int,
        PRIMARY KEY (id)
    )
    '''
)

#commit our changes
con.commit()


#close the connections
con.close()
```

```python
#Inserting our data into the new table

con = psycopg2.connect("dbname=national_parks_biodiversity user=postgres")
id_num = 100000
for i in plant_species_df.index:
    query = '''
    INSERT into plant_species_details8 (id, Park_name, Scientific_name, Category, Plant_Order, Family,
    Record_status, plant_animal_boolean) values(%s,'%s',%s,'%s','%s','%s','%s',%s);
    ''' % (id_num, plant_species_df['Park_name'][i], repr(plant_species_df['Scientific_name'][i]), plant_species_df['Category'][i],
          plant_species_df['Order'][i], plant_species_df['Family'][i],
          plant_species_df['Record_status'][i], plant_species_df['plant_animal_boolean'][i])
    id_num = id_num + 1
    cursor = con.cursor()
    print(query)
    cursor.execute(query)
    con.commit()
    cursor.close()
    #single_insert(conn, query)
# Close the connection
con.close()
```

```python
#Checking that we have the desired data stored in the table

con = psycopg2.connect("dbname=national_parks_biodiversity user=postgres")
cur = con.cursor()


cur.execute('SELECT * FROM plant_species_details')

all = cur.fetchall()

print(all)
```
