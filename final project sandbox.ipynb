```python
#imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#csv -> dataframe
species_df = pd.read_csv('species.csv', dtype = str)
parks_df = pd.read_csv('parks.csv')
```

```python
#drop that extra column in species_df
species_df = species_df.iloc[: , :-1]
species_df
```

```python
#dealing with NaNs
NaN_species_df = species_df.fillna(0)
NaN_species_df.head()
```

```python
#drop "in review" records
species_df_filtered = NaN_species_df[NaN_species_df['Record Status'] != "In Review"]
species_df_filtered
```

```python
#how many labels are in the Nativeness column + what are the labels?
species_df_filtered["Nativeness"].value_counts()
```

```python
#thinking that Nativeness = "Present" and "Not Confirmed" rows are not useful/incorrect
species_df_filt = species_df_filtered.drop(species_df_filtered[(species_df_filtered['Nativeness'] == "Present") | 
                                                               (species_df_filtered['Nativeness'] == "Not Confirmed")].index)
species_df_filt
```

```python
#check Natveness labels again
species_df_filt["Nativeness"].value_counts()
```

```python
#label encode Nativeness
from sklearn.preprocessing import LabelEncoder
class_labels = LabelEncoder()
species_df_filt['Nativeness']=class_labels.fit_transform(species_df_filt['Nativeness'].values)
species_df_filt.head()
```
