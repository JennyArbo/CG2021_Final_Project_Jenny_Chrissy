```python
#imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#csv -> dataframe
species_df = pd.read_csv('species.csv', dtype = str)
parks_df = pd.read_csv('parks.csv')
```

```python
species_df.head()
#Note: not sure why there is a 14th data column showing-there are only 13 columns of data-this is confirmed
#in the documentation for the dataset. Will come back and drop that.
```

```python
parks_df.head()
```

```python
#drop that extra column in species_df
species_df = species_df.iloc[: , :-1]
species_df
```

```python
#what are the possible labels for the "Record Status" column?
species_df["Record Status"].value_counts()
```

```python
#entries where "Record Status" is a Common Name clearly have errors
#entries with "In Review" record status are missing data in some fields,
#may not be as reliable as "Approved" entries

#keep only "Approved" entries
species_df_app = species_df[species_df['Record Status'] == "Approved"]
species_df_app
```

```python
#this data is primarily text
#so we will need to do some label encoding before we can create and train data models

#label encode Park Name
from sklearn.preprocessing import LabelEncoder
class_labels = LabelEncoder()
species_df_app['Park Name']=class_labels.fit_transform(species_df_app['Park Name'].values)
species_df_app
```

```python
#label encode Category
species_df_app['Category']=class_labels.fit_transform(species_df_app['Category'].values)
species_df_app
```

```python
#label encode Order
species_df_app['Order']=class_labels.fit_transform(species_df_app['Order'].values)
species_df_app
```

```python
#label encode Family
species_df_app['Family']=class_labels.fit_transform(species_df_app['Family'].values)
species_df_app
```

```python
#label encode Occurrence
species_df_app['Occurrence']=class_labels.fit_transform(species_df_app['Occurrence'].values)
species_df_app
```

```python
#label encode Abundance
species_df_app['Abundance']=class_labels.fit_transform(species_df_app['Abundance'].values)
species_df_app
```

```python
#label encode Seasonality
species_df_app['Seasonality']=class_labels.fit_transform(species_df_app['Seasonality'].values)
species_df_app
```

```python
#label encode Conservation Status
species_df_app['Conservation Status']=class_labels.fit_transform(species_df_app['Conservation Status'].values)
species_df_app
```

```python
#label encode Species ID
species_df_app['Species ID']=class_labels.fit_transform(species_df_app['Species ID'].values)
species_df_app
```

```python
#scientific names should be unique - each entry already has unique identifier in "Species ID"
#common names are useful for humans but may just muddy the waters for machine learning/modeling
#all remaining entries should have Record Status of "Approved" - this column can also be dropped

species_df_filtered = species_df_app.drop(labels = ["Scientific Name", "Common Names", "Record Status"], axis = 1)
```

```python
species_df_filtered.head()
```

```python
#how many labels are in the Nativeness column + what are the labels?
species_df_filtered["Nativeness"].value_counts()
```

```python
#label encode Nativeness
from sklearn.preprocessing import LabelEncoder
class_labels = LabelEncoder()
species_df_filtered['Nativeness']=class_labels.fit_transform(species_df_filtered['Nativeness'].values)
species_df_filtered.head()
```

```python
#check data types - see if our label encoding behaved
species_df_filtered.info()
```

```python
#import train_test_split
from sklearn.model_selection import train_test_split

X = species_df_filtered.drop('Nativeness', axis=1)
y = species_df_filtered['Nativeness']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=42)
```

```python
#scale dataset
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

```python
#set up Logistic Regression model
from sklearn.linear_model import LogisticRegression

regression = LogisticRegression()
regression.fit(X_train, y_train)
```

```python
#check performance of regression model
regression.fit(X_train, y_train)
y_pred = regression.predict(X_test)
regression.score(X_test, y_test)
```
